#Scraping all Game related articles from various websites and assigning the label "Game"

We created our custom dataset by scraping data from open websites (gamespot.com, ign.com, topgear.com, and autoblog.com) using the newspaper3k library, as listed below. We ran these cells once to create a csv file, which we will use for further processing.

```python
import sqlite3
import newspaper
import random
import csv

conn = sqlite3.connect('game_articles.db')
c = conn.cursor()

c.execute('CREATE TABLE IF NOT EXISTS articles (id INTEGER PRIMARY KEY AUTOINCREMENT, title TEXT, text TEXT, label TEXT)')


websites = ['https://store.steampowered.com/', 'https://www.ign.com/',
            'https://www.gamespot.com/', 'https://www.theverge.com/games',
            'https://en.wikipedia.org/wiki/Game']

for website in websites:
    print(f'Scraping {website}...')
    paper = newspaper.build(website)
    count = 0
    for article in paper.articles:
        if count == 200:
            break
        try:
            article.download()
            article.parse()
            title = article.title
            text = article.text
            c.execute("INSERT INTO articles (title, text, label) VALUES (?, ?, ?)", (title, text, 'game'))
            count += 1
        except:
            pass

conn.commit()

c.execute("SELECT title, text, label FROM articles WHERE label='game'")
results = c.fetchall()

random.shuffle(results)

with open('game_dataset.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['text', 'label'])
    for result in results:
        writer.writerow([result[1], 'game'])

conn.close()
```

# Scraping all car related articles from various websites and assigning the label "car"

```python
import sqlite3
import newspaper
import random
import csv

conn = sqlite3.connect('car_articles.db')
c = conn.cursor()

c.execute('CREATE TABLE IF NOT EXISTS articles (id INTEGER PRIMARY KEY AUTOINCREMENT, title TEXT, text TEXT, label TEXT)')

websites = ['https://www.caranddriver.com/news/', 'https://www.motortrend.com/', 'https://www.autoblog.com/',
            'https://www.topgear.com/car-news', 'https://www.cnn.com/business/cars']

for website in websites:
    print(f'Scraping {website}...')
    paper = newspaper.build(website)
    count = 0
    for article in paper.articles:
        if count == 200:
            break
        try:
            article.download()
            article.parse()
            title = article.title
            text = article.text
            c.execute("INSERT INTO articles (title, text, label) VALUES (?, ?, ?)", (title, text, 'car'))
            count += 1
        except:
            pass

conn.commit()
c.execute("SELECT title, text, label FROM articles WHERE label='car'")
results = c.fetchall()

random.shuffle(results)

with open('car_dataset.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['text', 'label'])
    for result in results:
        writer.writerow([result[1], 'car'])

conn.close()
```

#Keeping only English articles

```python
import langdetect

def detect_language(text):
    try:
        return langdetect.detect(text)
    except:
        return 'unknown'

df_combined['language'] = df_combined['text'].apply(detect_language)
```

```python
df_combined.label.value_counts()
```

```python
df_combined = df_combined[df_combined['language'] == 'en'].iloc[:,:2]
```

Randomly choose two sentences from each article to reduce the data

```python
import random

def get2sentences(row):
    if row == "":
        return row
    row = row.replace('\n',".")
    row = row.replace('!',".")
    sentences = row.split(".")
    random.shuffle(sentences)
    if len(sentences)>4:
        return ".".join(sentences[:2])+"."
    return row
```

```python
df_combined = df_combined.reset_index(drop=True)
df_combined['text'] = df_combined['text'].apply(get2sentences)
```

#Splitting the data into train and test samples

We have split the dataset into train‑test ratio of 0.2 which means there are at least 190+ samples of

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_combined.text, df_combined.label, test_size=0.2, random_state=23)
```

#Tokenize the data using the AutoTokenizer from "bert‑base‑cased" model.

Since the BERT-base-cased model has been trained on a vast and varied corpus of text data, which includes language related to both games and cars, it is effective at predicting labels for sentences pertaining to the comparison of games and cars. The model's comprehension of pertinent vocabulary and semantic structures can be utilized to make precise predictions when it is fine-tuned for a task of classifying sentences related to games versus cars.

```python
!pip install transformers
```

```python
from transformers import AutoTokenizer
from transformers import TFAutoModelForSequenceClassification
from tensorflow.keras.optimizers import Adam
import numpy

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
def tokenize(X):
    #truncation for not exceeding 512 limit
    tokenized_data = tokenizer(X, return_tensors="np", truncation=True, padding=True)
    tokenized_data = dict(tokenized_data)
    return tokenized_data
```

```python
X_train = tokenize(list(X_train.astype(str)))
X_test = tokenize(list(X_test.astype(str)))

from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
y_train= label_encoder.fit_transform(y_train)

label_encoder = preprocessing.LabelEncoder()
y_test= label_encoder.fit_transform(y_test)
```

```python
model = TFAutoModelForSequenceClassification.from_pretrained("bert-base-cased")
model.compile(optimizer=Adam(3e-5))
model.fit(X_train, y_train)
```

```python
y_pr = model.predict(X_test)["logits"]
y_pred = np.argmax(y_pr, axis=1)

from sklearn.metrics import accuracy_score, confusion_matrix
confusion_matrix(y_true=y_test, y_pred=y_pred)
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
```

We obtain a test accuracy of 81.3% which is pretty good for random data scraped from the internet which was a part of articles related to the 2 labels, car and game.
